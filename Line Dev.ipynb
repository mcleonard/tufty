```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import tufty as tft
```

```python
x = np.linspace(0, 10, num=500)
y1 = np.sin(x) + np.random.randn(len(x))
y2 = np.sin(2*(x+1)) + np.random.randn(len(x))*0.5
df = pd.DataFrame({'y1':y1, 'y2':y2}, index=x)
```

```python
plt.plot(x, df[['y1', 'y2']])
```

## Simple Line Plot

```python
plot = tft.Plot(title="Signals").line(x, y1)
plot.line(x, y2)
plot.labels(x='Time (us)', y='Voltage (V)');
```

## DataFrame Line Plot

```python
plot = tft.Plot(df, title="Signals").line(x='index', y='y1')
plot.line(x='index', y='y2')
plot.labels(x='Time (us)', y='Voltage (V)');
```

```python
tft.Plot(df, title="Signals").line(x='index', y=['y1', 'y2'])
```
